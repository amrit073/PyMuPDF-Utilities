# Notebook: Demonstration of PyMuPDF's new Table Feature

```python
import fitz  # import PyMuPDF
from show_image import show_image  # import image display feature
```

# Open example document and show the first page

```python
doc = fitz.open("chinese-table.pdf")
page = doc[0]
page.clean_contents()
```

```python
for i,tab in enumerate(page.find_tables()):
    for cell in tab.header.cells:
        page.draw_rect(cell,color=fitz.pdfcolor["red"],width=0.3)
    page.draw_rect(tab.bbox,color=fitz.pdfcolor["green"])
    print(f"Table {i}:", tab.header.names, f"external: {tab.header.external}")
    for l in tab.extract():
        print(l)
    print("-"*50)
    
show_image(page, f"Table & Header")
```
